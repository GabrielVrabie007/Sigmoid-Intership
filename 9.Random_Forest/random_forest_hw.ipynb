## **Decision Boundary Graph** 

```python
def plot_decision_boundaries(X, y, classifier, feature_names, model_name):
    plt.figure(figsize=(8, 8))
    
    colors = ['green', 'lime']
    cmap = ListedColormap(['darkorange', 'gold'])
    
    disp = DecisionBoundaryDisplay.from_estimator(
        classifier, X, response_method="predict",
        xlabel=feature_names[0], ylabel=feature_names[1],
        alpha=0.7, grid_resolution=100, cmap=cmap
    )
    
    scatter = disp.ax_.scatter(
        X[:, 0], X[:, 1], 
        c=y, cmap=ListedColormap(colors),
        s=120, edgecolor='k', linewidth=1.5, alpha=0.8
    )
    
    plt.title(model_name)
    legend_labels = {0: 'Outcome 0', 1: 'Outcome 1'}
    handles, _ = scatter.legend_elements()
    plt.legend(handles, legend_labels.values(), loc="upper left", title="Outcome")
    
    plt.tight_layout()
    plt.show()
```

### Training and testing models

```python

features = ["ST slope", "exercise angina"]
X = df[features].values 
y = df['target'].values
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

random_forest_model = RandomForestClassifier(random_state=42)
random_forest_model.fit(X_train_scaled,y_train)

decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train_scaled, y_train)

y_pred_random_forest = random_forest_model.predict(X_test_scaled)
y_pred_decision_tree = decision_tree_model.predict(X_test_scaled)
print("Accuracy Random Forest:", accuracy_score(y_test, y_pred_random_forest))
print("Accuracy Decision Tree:", accuracy_score(y_test, y_pred_decision_tree))
```

```python

plot_decision_boundaries(X_train_scaled, y_train, decision_tree_model, features, "Decision Tree")

plot_decision_boundaries(X_test_scaled, y_test, random_forest_model, features, "Random Forest")
```

## 💡How It Works

### Building the Forest
1. **Create multiple trees** (typically 100 or more)
2. For each tree:
   - **Bootstrap sampling**: Randomly select rows from the dataset with replacement
     - Example: From 1000 rows, sample 100 rows
   - **Feature randomization**: At each split, consider only a subset of features
     - Example: Maybe only ['ST slope', 'oldpeak', 'sex'] from all available features
   - The tree learns decision rules like: `if ST slope > 1.5 and angina = yes, then target = 1`

### 🔍 Making Predictions
1. **Collect votes** from all trees for a new patient's data
   - Each tree independently predicts target = 0 or target = 1
2. **Majority voting**: Count all votes and return the most common prediction
   - Example: If 70 trees predict class 1 and 30 trees predict class 0, the final prediction is class 1


The diagram shows how Random Forest creates more complex decision boundaries by combining multiple trees, compared to a single decision tree.

#### Association learning is one of the best methods, so imagine that a ***random forest is like a parliament*** where each ***tree is a deputy***. Each one votes, then the calculation is made and the result is determined.

#### This is a great analogy for understanding how Random Forest works! It helps visualize how multiple decision trees (the "deputies") independently cast their votes, and then the majority decision becomes the final prediction of the model.

# 💡 Studying Hyperparameters

```python
def plot_hyperparameter_effect(param_name, param_range, X_train, X_test, y_train, y_test, ax, random_state=42):
    test_scores = []
    
    for value in param_range:
        params = {param_name: value, 'random_state': random_state}
        model = RandomForestClassifier(**params)
        model.fit(X_train, y_train)
        
        # Accuracy only for test set
        test_scores.append(model.score(X_test, y_test))
    
    x_labels = [str(val) if val is not None else "None" for val in param_range]
    
    ax.plot(range(len(param_range)), test_scores, 'o-', color='orange', linewidth=2)
    
    # Find best value
    best_idx = np.argmax(test_scores)
    best_value = param_range[best_idx]
    best_score = test_scores[best_idx]
    
    # Mark in plot best value with red star
    ax.plot(best_idx, best_score, 'r*', markersize=12)
    
    ax.set_title(f'Effect of {param_name}')
    ax.set_xlabel(param_name)
    ax.set_ylabel('Test Accuracy')
    ax.set_xticks(range(len(param_range))[::max(1, len(param_range)//5)])
    ax.set_xticklabels([x_labels[i] for i in range(len(param_range))[::max(1, len(param_range)//5)]], rotation=45)
    ax.grid(True, linestyle='--', alpha=0.5)
    
    return best_value, best_score
```

# Key RandomForestClassifier Parameters

| Parameter | Type | Description |
|-----------|------|-------------|
| `n_estimators` | integer | Number of trees in the forest. Higher values (100-500) generally improve accuracy but increase training time. |
| `max_depth` | integer or None | Maximum depth of each tree. Controls overfitting. Typical production values: 10-30. None allows unlimited growth. |
| `min_samples_split` | integer or float | Minimum samples needed to split a node. Higher values (5-10) prevent splits based on noise. |
| `min_samples_leaf` | integer or float | Minimum samples required in leaf nodes. Effective for preventing overfitting. Values 1-10 depending on dataset size. |
| `max_features` | string/int/float | Features considered for best split. Options: 'sqrt', 'log2', int, or float. Controls tree diversity. |
| `bootstrap` | boolean | Whether to use bootstrap sampling. True enables the "random" in Random Forest by creating diverse trees. |
| `random_state` | integer | Controls randomness. Essential for reproducible results in production environments. |
| `n_jobs` | integer | Number of parallel jobs. Set to -1 to use all processors and speed up training. |
| `class_weight` | dict/'balanced' | Adjusts for class imbalance. 'balanced' option automatically weights classes inversely to frequencies. |

```python
hyperparameters = {
    'max_depth': list(range(1, 21)),
    'min_samples_split': list(range(2, 21)),
    'min_samples_leaf': list(range(1, 21)),
    'min_weight_fraction_leaf': [0.0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5],
    'max_features': [None, 'sqrt', 'log2'] + list(range(1, X_train.shape[1] + 1)),
    'max_leaf_nodes': [None] + list(range(2, 21)),
    'min_impurity_decrease': [0.0, 0.001, 0.002, 0.005, 0.01, 0.02, 0.05, 0.1, 0.2, 0.5]
}

fig, axes = plt.subplots(4, 2, figsize=(15, 18))
axes = axes.flatten()

best_params = {}

# Plot for each hyperparameter
for i, (param_name, param_range) in enumerate(hyperparameters.items()):
    if i < len(axes):
        best_value, best_score = plot_hyperparameter_effect(
            param_name, param_range, X_train, X_test, y_train, y_test, axes[i])
        best_params[param_name] = (best_value, best_score)
    
if len(hyperparameters) < len(axes):
    axes[-1].axis('off')

plt.suptitle('Effect of Random Forest Classifier Hyperparameters on Test Accuracy', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
print("Best hyperparameter values:")
for param_name, (best_value, best_score) in best_params.items():
    print(f"{param_name}: {best_value}")
```

### Training and testing model with tunned hyperparameters

```python
# best_params in my case a dictionary with tuples as values so I need to extract the first value of each tuple
first_values = {k: v[0] if isinstance(v, tuple) else v for k, v in best_params.items()}

# Default parameters
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Default parameters accuracy:", accuracy_score(y_test, y_pred))

# Best parameters
best_clf = RandomForestClassifier(random_state=42, **first_values)
best_clf.fit(X_train, y_train)

y_pred = best_clf.predict(X_test)
print("Best parameters accuracy:", accuracy_score(y_test, y_pred))
```
